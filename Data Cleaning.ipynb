```python
def convert_to_snake_case(df):
    df.columns = [
        inflection.underscore(
            col.strip().replace("&", "").replace("(", "").replace(")", "").replace("-", "").replace(" ", "_").replace("__", "_")
        )
        for col in df.columns
    ]
    return df
```

```python
def calculate_null_percentage(df):
   
    total_count = len(df)
    df_info = pd.DataFrame({
        "Column": df.columns,
        "Non-Null Count": df.notnull().sum(),
        "Total Count": total_count,
    })
    
    df_info['Null Count'] = df_info['Total Count'] - df_info['Non-Null Count']
    df_info['Null Percentage'] = (df_info['Null Count'] / df_info['Total Count']) * 100
    df_info_sorted = df_info[['Column', 'Null Percentage']].sort_values(by='Null Percentage', ascending=False)
    return df_info_sorted
```

```python

benzine = ["Gasoline", "Super 95", 'Regular/Benzine 91', "Super E10 95", "Regular/Benzine 91 (Particle filter)", 
           "Super 95 (Particle filter)", "Super E10 95 (Particle filter)", "Regular/Benzine E10 91 ", "Super Plus 98",
           "Super Plus E10 98", "Super Plus E10 98 (Particle filter)", 'Others', "Others (Particle filter)", 
           "Super Plus 98 (Particle filter)", "Regular/Benzine E10 91 (Particle filter)", "Regular/Benzine E10 91", 
           "Gasoline (Particle filter)", "Super 95 (Particle filter) / Super E10 95 / Ethanol", "Ethanol"]
lpg = ["Liquid petroleum gas (LPG)", "Domestic gas L", "CNG (Particle filter)", "LPG (Particle filter) ","Liquid petroleum gas",
       "Liquid petroleum gas (LPG) (Particle filter)","LPG", "CNG", "Biogas", "Domestic gas H", "LPG (Particle filter)",
       "Domestic gas L (Particle filter)", "Hydrogen"]

lpg_benzine = ['Liquid petroleum gas (LPG) (Particle filter) / Super 95 / Super E10 95','Liquid petroleum gas (LPG) / Super 95 / Super E10 95',
              'Liquid petroleum gas (LPG) / Super 95 / Super Plus 98 / Super Plus E10 98 / Super E10 95', 'Liquid petroleum gas (LPG) / Super 95',
              'Liquid petroleum gas (LPG) / Super 95 / Super E10 95 / Super Plus 98', 'Liquid petroleum gas (LPG) / Super E10 95', 
              'Liquid petroleum gas (LPG) / Super E10 95 / Super Plus E10 98 / Super Plus 98 / Super 95', 
               'Domestic gas H / Super E10 95 / Super Plus E10 98 / Super 95 / Super Plus 98 / Domestic gas L',
               'Liquid petroleum gas (LPG) / Super 95 / Super Plus 98', 'Super 95 / Super Plus 98 / Liquid petroleum gas (LPG)',
               'Domestic gas L / Super 95 / Domestic gas H', 
               'Liquid petroleum gas (LPG) / Super E10 95 / Regular/Benzine 91 / Super 95 / Super Plus 98 / Biogas',
               'Super 95 / Liquid petroleum gas (LPG)']

diesel = ['Diesel', 'Diesel (Particle filter)']

electric = ['Electric', 'Electric (Particle filter)']

def fueltype(x):
    if x in benzine:
        return "Benzine"
    elif x in lpg:
        return 'LPG/CNG'
    elif x in lpg_benzine:
        return 'LPG/Benzine'
    elif x in diesel:
        return 'Diesel'
    elif x in electric:
        return 'Electric'
    else:
        return x
```
